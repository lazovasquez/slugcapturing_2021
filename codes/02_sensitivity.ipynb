## Sensitivity figure

```python
import matplotlib.pyplot as plt
import numpy as np

# Some example data to display
x = np.linspace(0, 2 * np.pi, 400)
y = np.sin(x ** 2)

fig, axs = plt.subplots(4, 2)

mapsize = [12, 8]
plt.rcParams ['figure.figsize'] = mapsize
plt.grid (True, which = "both")

axs[0, 0].plot(x, y)
axs[0, 0].set_title('Continuous Galerkin')
axs[0, 1].plot(x, y, 'tab:orange')
axs[0, 1].set_title('Taylor Hood')

axs[1, 0].plot(x, -y, 'tab:green')
axs[1, 0].set_title('p = 2')
axs[1, 1].plot(x, -y, 'tab:red')
axs[1, 1].set_title('p = 3')

axs[2, 0].plot(x, -y, 'tab:green')
axs[2, 0].set_title('nx = 40')
axs[2, 1].plot(x, -y, 'tab:red')
axs[2, 1].set_title('nx = 400')

axs[3, 0].plot(x, -y, 'tab:green')
axs[3, 0].set_title('nx = 40')
axs[3, 1].plot(x, -y, 'tab:red')
axs[3, 1].set_title('nx = 400')

for ax in axs.flat:
    ax.set(xlabel='x-label', ylabel='y-label')

# Hide x labels and tick labels for top plots and y ticks for right plots.
for ax in axs.flat:
    ax.label_outer()
```
