```python
"""
Two fluid model
compressible
differential equation system
Fourier analysis
"""
j_lvector       = np.linspace (minj_l, maxj_l, num = testpointsj_l)
j_gvector       = np.linspace (minj_g, maxj_g, num = testpointsj_g)
j_lmapstability = np.zeros ((testpointsbeta,testpointsj_l)) #(rows,columns)

k = 0
for beta in betavector:
    print ("  >>> β = ", np.rad2deg(beta), "[°]")
    j = 0 #for each element of j2vector
    for j_g in j_gvector: #iterative process for j2
        i = 0 #for each element of j1vector
        print(" >> j_g = ", j_g)
        for j_l in j_lvector: #iterative process for j1
            print(" > j_l = ", j_l)
            ############
            # VARIABLES
            # var1: liquid volume fraction (alpha_l)
            # var2: liquid velocity (u_l)
            # var3: gas velocity (u_g)
            # var4: interfacial pressure (p_i)

            # FUNCTIONS
            # stratification angle
            def gamma (var1):
                return pi*var1 + pow ((3.*pi/2.), (1./3.))*(1 - 2*var1 + pow (abs (var1), (1./3.)) - pow (abs(1 - var1), (1/3))) - (1/200)*(1-var1)*var1*(1 - 2*var1)*(1+4*((1-var1)**2+(var1)**2))

            # Gas density
            def rho_g (var4):
                return p_factor*var4

            # liquid phase velocity
            def var2 (var1):
                return j_l/(var1)

            # gas phase velocity
            def var3 (var1):
                return j_g/(1 - var1)

            # liquid sectional area
            def A_l (var1):
                return A*var1

            # gas sectional area
            def A_g (var1):
                return A*(1 - var1)

            # liquid wetted perimeter
            def P_lw (var1):
                return D*gamma (var1)

            # gas wetted perimeter
            def P_gw (var1):
                return D*(pi - gamma (var1))

            # interface wetted perimeter
            def P_gl (var1):
                return D*np.sin (gamma (var1))

            def P_gl_fenics (var1):
                return D*sin (gamma (var1))

            # liquid hydraulic diameter 
            def Dh_l (var1):
                return 4.*A_l (var1)/P_lw (var1)

            # gas hydraulic diameter 
            def Dh_g (var1):
                return 4.*A_g (var1)/(P_gw (var1) + P_gl (var1))

            def Dh_g_fenics (var1):
                return 4.*A_g (var1)/(P_gw (var1) + P_gl_fenics (var1))

            # interface hydraulic diameter 
            def Dh_gl (var1):
                return 4.*A_g (var1)/(P_gw (var1) + P_gl (var1))

            def Dh_gl_fenics (var1):
                return 4.*A_g (var1)/(P_gw (var1) + P_gl_fenics (var1))

            # critic diameter
            def Dc (var1):
                return g*np.cos (beta)*pi*D/(4*np.sin (gamma (var1)))

            def Dc_fenics (var1):
                return g*cos (beta)*pi*D/(4*sin (gamma (var1)))

            # liquid Reynolds number
            def Re_l (var1, var2):
                return rho_l*var2*Dh_l (var1)/mu_l

            # def Re_l (var1, var2):
            #     return D*j_l*rho_l/mu_l
            # def Re_l (var1, var2):  #ppppp
            #     return D*var1*var2*rho_l/mu_l

            # gas Reynolds number
            def Re_g (var1, var3, var4):
                return rho_g (var4)*var3*Dh_g (var1)/mu_g

            def Re_g_fenics (var1, var3, var4):
                return rho_g (var4)*var3*Dh_g_fenics (var1)/mu_g

            # interface Reynolds number
            def Re_gl (var1, var2, var3, var4):
                return rho_g (var4)*abs(var3 - var2)*Dh_gl (var1)/mu_g

            def Re_gl_fenics (var1, var2, var3, var4):
                return rho_g (var4)*abs(var3 - var2)*Dh_gl_fenics (var1)/mu_g

            # liquid friction factor
            # def f_lw (var1, var2):
            #     Af_lw = pow (2.457*np.log (pow (pow (abs (7/Re_l (var1, var2)), 0.9) + 0.27*roughness/Dh_l (var1), -1)), 16)
            #     Bf_lw = pow (abs (37530/Re_l (var1, var2)), 16)
            #     return 2*pow (pow (abs (8/Re_l (var1, var2)), 12) + pow (abs (Af_lw + Bf_lw), - 1.5), 1/12)

            def f_lw_fenics (var1, var2):
                Af_lw = pow (2.457*ln (pow (pow (abs (7/Re_l (var1, var2)), 0.9) + 0.27*roughness/Dh_l (var1), -1)), 16)
                Bf_lw = pow (abs (37530/Re_l (var1, var2)), 16)
                return 2*pow (pow (abs (8/Re_l (var1, var2)), 12) + pow (abs (Af_lw + Bf_lw), - 1.5), 1/12)

            def f_lw (var1, var2):
                return 0.046*abs(Re_l (var1, var2))**(-0.2)

            # def f_lw (var1, var2):
            #     return 0.0262*pow (abs(var1*Re_l (var1, var2)), -0.139)

            def f_lw_laminar (var1, var2):
                return 24/Re_l (var1, var2)

            # gas friction factor
            # def f_gw (var1, var3, var4):
            #     Af_gw = pow (2.457*np.log (pow (pow (abs (7/Re_g (var1, var3, var4)), 0.9) + 0.27*roughness/Dh_g (var1), - 1)), 16)
            #     Bf_gw = pow (abs (37530/Re_g (var1, var3, var4)), 16)
            #     return 2*pow (pow (abs (8/Re_g (var1, var3, var4)), 12) + pow (abs (Af_gw + Bf_gw), - 1.5), 1/12)

            def f_gw_fenics (var1, var3, var4):
                Af_gw = pow (2.457*ln (pow (pow (abs (7/Re_g_fenics (var1, var3, var4)), 0.9) + 0.27*roughness/Dh_g_fenics (var1), - 1)), 16)
                Bf_gw = pow (abs (37530/Re_g_fenics (var1, var3, var4)), 16)
                return 2*pow (pow (abs (8/Re_g_fenics (var1, var3, var4)), 12) + pow (abs (Af_gw + Bf_gw), - 1.5), 1/12)

            def f_gw (var1, var3, var4):
                return 0.046*abs(Re_g (var1, var3, var4))**(-0.2)

            # def f_gw_fenics (var1, var3, var4):
            #     return 0.046*pow (abs(Re_g_fenics (var1, var3, var4)), -0.2)

            def f_gw_laminar (var1, var3, var4):
                return 16 / Re_g (var1, var3, var4)

            # interfacial friction
            def f_gl (var1, var2, var3, var4):
                return max (f_gw (var1, var3, var4), 0.014)

            # def f_gl (var1, var2, var3, var4):
            #     return 0.046*pow (abs(Re_gl (var1, var2, var3, var4)), -0.2)
            def f_gl_laminar (var1, var2, var3, var4):
                return 16 / Re_gl (var1, var2, var3, var4)
            # def f_gl_fenics (var1, var2, var3, var4):
            #     return 0.046*pow (abs(Re_gl_fenics (var1, var2, var3, var4)), -0.2)
            def f_gl_fenics (var1, var2, var3, var4):
                return max (f_gw_fenics (var1, var3, var4), 0.014)
            # liquid shear stress
            def tau_lw (var1, var2):
                return 1/2*f_lw (var1, var2)*rho_l*var2*abs (var2)
            # gas shear stress
            def tau_gw (var1, var3, var4):
                return 1/2*f_gw (var1, var3, var4)*rho_g (var4)*var3*abs (var3)
            def tau_gw_fenics (var1, var3, var4):
                return 1/2*f_gw_fenics (var1, var3, var4)*rho_g (var4)*var3*abs (var3)
            # interface shear stress l
            def tau_gl (var1, var2, var3, var4):
                return 1/2*f_gl(var1, var2, var3, var4)*rho_g (var4)*(var3 - var2)*abs (var3 - var2)

            def tau_gl_fenics (var1, var2, var3, var4):
                return 1/2*f_gl_fenics (var1, var2, var3, var4)*rho_g (var4)*(var3 - var2)*abs (var3 - var2)
            # steady state momentum equation for horizontal steady state (characteristics and fourier). var4 is known, compute var1
            def equilibrium1 (var1): 
                return (rho_g (var4) - rho_l)*g*np.sin (beta) - tau_lw (var1, var2 (var1))*P_lw (var1)/A_l (var1) + tau_gw (var1, var3 (var1), var4)*P_gw (var1)/A_g (var1) + tau_gl (var1, var2 (var1), var3 (var1), var4)*P_gl (var1)*(1/A_l (var1) + 1/A_g (var1))
            # liquid phase velocity
            def var2b (j_l):
                return j_l/(var1)
            # stady state momentum equation for horizontal steady state (characteristics and fourier). var4 is known, compute j_l
            def equilibrium2 (j_l):  
                return (rho_g (var4) - rho_l)*g*np.sin (beta) - tau_lw (var1, var2b (j_l))*P_lw (var1)/A_l (var1) + tau_gw (var1, var3 (var1), var4)*P_gw (var1)/A_g (var1) + tau_gl (var1, var2b (j_l), var3 (var1), var4)*P_gl (var1)*(1/A_l (var1) + 1/A_g (var1))

            # Amat
            def Amat (var1, var2, var3, var4):
                Amat = [[A11 (var1 ,var2, var3, var4), A12 (var1 ,var2, var3, var4), A13 (var1 ,var2, var3, var4), A14 (var1 ,var2, var3, var4)], 

                [A21 (var1 ,var2, var3, var4), A22 (var1 ,var2, var3, var4), A23 (var1 ,var2, var3, var4), A24 (var1 ,var2, var3, var4)], 

                [A31 (var1 ,var2, var3, var4), A32 (var1 ,var2, var3, var4), A33 (var1 ,var2, var3, var4), A34 (var1 ,var2, var3, var4)], 

                [A41 (var1 ,var2, var3, var4), A42 (var1 ,var2, var3, var4), A43 (var1 ,var2, var3, var4), A44 (var1 ,var2, var3, var4)]]
                return Amat

            # Bmat
            def Bmat (var1, var2, var3, var4):
                Bmat = [[B11 (var1 ,var2, var3, var4), B12 (var1 ,var2, var3, var4), B13 (var1 ,var2, var3, var4), B14 (var1 ,var2, var3, var4)], 

                [B21 (var1 ,var2, var3, var4), B22 (var1 ,var2, var3, var4), B23 (var1 ,var2, var3, var4), B24 (var1 ,var2, var3, var4)], 

                [B31 (var1 ,var2, var3, var4), B32 (var1 ,var2, var3, var4), B33 (var1 ,var2, var3, var4), B34 (var1 ,var2, var3, var4)], 
                
                [B41 (var1 ,var2, var3, var4), B42 (var1 ,var2, var3, var4), B43 (var1 ,var2, var3, var4), B44 (var1 ,var2, var3, var4)]]
                return Bmat

            # Cmat linearization
            def Cmat1_var1 (var1):
                var2 = ref[1]
                var3 = ref[2]
                var4 = ref[3]
                return Cmat1 (var1, var2, var3, var4)
            def Cmat1_var2 (var2):
                var1 = ref[0]
                var3 = ref[2]
                var4 = ref[3]
                return Cmat1 (var1, var2, var3, var4)
            def Cmat1_var3 (var3):
                var1 = ref[0]
                var2 = ref[1]
                var4 = ref[3]
                return Cmat1 (var1, var2, var3, var4)
            def Cmat1_var4 (var4):
                var1 = ref[0]
                var2 = ref[1]
                var3 = ref[2]
                return Cmat1 (var1, var2, var3, var4)

            def Cmat2_var1 (var1):
                var2 = ref[1]
                var3 = ref[2]
                var4 = ref[3]
                return Cmat2 (var1, var2, var3, var4)
            def Cmat2_var2 (var2):
                var1 = ref[0]
                var3 = ref[2]
                var4 = ref[3]
                return Cmat2 (var1, var2, var3, var4)
            def Cmat2_var3 (var3):
                var1 = ref[0]
                var2 = ref[1]
                var4 = ref[3]
                return Cmat2 (var1, var2, var3, var4)
            def Cmat2_var4 (var4):
                var1 = ref[0]
                var2 = ref[1]
                var3 = ref[2]
                return Cmat2 (var1, var2, var3, var4)

            def Cmat3_var1 (var1):
                var2 = ref[1]
                var3 = ref[2]
                var4 = ref[3]
                return Cmat3 (var1, var2, var3, var4)
            def Cmat3_var2 (var2):
                var1 = ref[0]
                var3 = ref[2]
                var4 = ref[3]
                return Cmat3 (var1, var2, var3, var4)
            def Cmat3_var3 (var3):
                var1 = ref[0]
                var2 = ref[1]
                var4 = ref[3]
                return Cmat3 (var1, var2, var3, var4)
            def Cmat3_var4 (var4):
                var1 = ref[0]
                var2 = ref[1]
                var3 = ref[2]
                return Cmat3 (var1, var2, var3, var4)

            def Cmat4_var1 (var1):
                var2 = ref[1]
                var3 = ref[2]
                var4 = ref[3]
                return Cmat4(var1, var2, var3, var4)
            def Cmat4_var2 (var2):
                var1 = ref[0]
                var3 = ref[2]
                var4 = ref[3]
                return Cmat4(var1, var2, var3, var4)
            def Cmat4_var3 (var3):
                var1 = ref[0]
                var2 = ref[1]
                var4 = ref[3]
                return Cmat4(var1, var2, var3, var4)
            def Cmat4_var4 (var4):
                var1 = ref[0]
                var2 = ref[1]
                var3 = ref[2]
                return Cmat4(var1, var2, var3, var4)

            # Cmat_lin
            def Cmat_lin (var1, var2, var3, var4):
                Cmat_lin = [[gradient (Cmat1_var1, 1), gradient (Cmat1_var2, 2), gradient (Cmat1_var3, 3), gradient (Cmat1_var4, 4)], 
                            [gradient (Cmat2_var1, 1), gradient (Cmat2_var2, 2), gradient (Cmat2_var3, 3), gradient (Cmat2_var4, 4)], 
                            [gradient (Cmat3_var1, 1), gradient (Cmat3_var2, 2), gradient (Cmat3_var3, 3), gradient (Cmat3_var4, 4)], 
                            [gradient (Cmat4_var1, 1), gradient (Cmat4_var2, 2), gradient (Cmat4_var3, 3), gradient (Cmat4_var4, 4)]]
                return Cmat_lin

            # ===============================================================
            # 1 - QUASILINEAR SYSTEM 
            # W = (alpha_l, u_l, u_g, p_i)
            # A != I
            # ===============================================================
            # Matrices
            if system == "system_1":
                # Amat
                def A11 (var1 ,var2, var3, var4):
                    return rho_l
                def A12 (var1 ,var2, var3, var4):
                    return 0
                def A13 (var1 ,var2, var3, var4):
                    return 0
                def A14 (var1 ,var2, var3, var4):
                    return 0

                def A21 (var1 ,var2, var3, var4):
                    return -rho_g (var4)
                def A22 (var1 ,var2, var3, var4):
                    return 0
                def A23 (var1 ,var2, var3, var4):
                    return 0
                def A24 (var1 ,var2, var3, var4):
                    return p_factor*(1 - var1)

                def A31 (var1 ,var2, var3, var4):
                    return rho_l*var2
                def A32 (var1 ,var2, var3, var4):
                    return rho_l*var1
                def A33 (var1 ,var2, var3, var4):
                    return 0
                def A34 (var1 ,var2, var3, var4):
                    return 0

                def A41 (var1 ,var2, var3, var4):
                    return -rho_g (var4)*var3
                def A42 (var1 ,var2, var3, var4):
                    return 0
                def A43 (var1 ,var2, var3, var4):
                    return rho_g (var4)*(1 - var1)
                def A44 (var1 ,var2, var3, var4):
                    return p_factor*(1 - var1)*var3

                # Bmat
                def B11 (var1 ,var2, var3, var4):
                    return rho_l*var2
                def B12 (var1 ,var2, var3, var4):
                    return rho_l*var1
                def B13 (var1 ,var2, var3, var4):
                    return 0
                def B14 (var1 ,var2, var3, var4):
                    return 0

                def B21 (var1 ,var2, var3, var4):
                    return -rho_g (var4)*var3
                def B22 (var1 ,var2, var3, var4):
                    return 0
                def B23 (var1 ,var2, var3, var4):
                    return rho_g (var4)*(1 - var1)
                def B24 (var1 ,var2, var3, var4):
                    return p_factor*(1 - var1)*var3

                def B31 (var1 ,var2, var3, var4):
                    return rho_l*var2**2 + rho_l*Dc (var1)*var1
                def B32 (var1 ,var2, var3, var4):
                    return 2*rho_l*var1*var2
                def B33 (var1 ,var2, var3, var4):
                    return 0
                def B34 (var1 ,var2, var3, var4):
                    return var1

                def B41 (var1 ,var2, var3, var4):
                    return -rho_g (var4)*var3**2 + rho_g (var4)*Dc (var1)*(1 - var1)
                def B42 (var1 ,var2, var3, var4):
                    return 0
                def B43 (var1 ,var2, var3, var4):
                    return rho_g (var4)*2*(1 - var1)*var3
                def B44 (var1 ,var2, var3, var4):
                    return p_factor*(1 - var1)*var3**2 + (1 - var1)

                # Cmat
                def Cmat1 (var1, var2, var3, var4):
                    return 0
                def Cmat2 (var1, var2, var3, var4):
                    return 0
                def Cmat3 (var1, var2, var3, var4):
                    return -rho_l*g*np.sin (beta)*var1 - tau_lw (var1, var2)*P_lw (var1)/A + tau_gl (var1, var2, var3, var4)*P_gl (var1)/A + Fbody*var1
                def Cmat4 (var1, var2, var3, var4):
                    return -rho_g (var4)*g*np.sin (beta)*(1 - var1) - tau_gw (var1, var3, var4)*P_gw (var1)/A - tau_gl (var1, var2, var3, var4)*P_gl (var1)/A + Fbody*(1 - var1)

                # Dmat
                def Dmat (var1, var4, Gamma_l, Gamma_g, nu_l, nu_g):
                    Dmat = [[rho_l*Gamma_l, 0, 0, 0],
                    
                            [-rho_g (var4)*Gamma_g, 0, 0, 0],
                            
                            [0, var1*nu_l, 0, 0],
                            
                            [0, 0, nu_g*(1 - var1), 0]]
                    return Dmat

            # ===============================================================
            # 2 - QUASILINEAR SYSTEM 
            # W = (alpha_l, u_l, u_g, p_i)
            # A = I 
            # ===============================================================
            # Matrices
            elif system == "system_2":
                # Amat
                def A11 (var1 ,var2, var3, var4):
                    return 1
                def A12 (var1 ,var2, var3, var4):
                    return 0
                def A13 (var1 ,var2, var3, var4):
                    return 0
                def A14 (var1 ,var2, var3, var4):
                    return 0

                def A21 (var1 ,var2, var3, var4):
                    return 0
                def A22 (var1 ,var2, var3, var4):
                    return 1
                def A23 (var1 ,var2, var3, var4):
                    return 0
                def A24 (var1 ,var2, var3, var4):
                    return 0

                def A31 (var1 ,var2, var3, var4):
                    return 0
                def A32 (var1 ,var2, var3, var4):
                    return 0
                def A33 (var1 ,var2, var3, var4):
                    return 1
                def A34 (var1 ,var2, var3, var4):
                    return 0

                def A41 (var1 ,var2, var3, var4):
                    return 0
                def A42 (var1 ,var2, var3, var4):
                    return 0
                def A43 (var1 ,var2, var3, var4):
                    return 0
                def A44 (var1 ,var2, var3, var4):
                    return 1

                # Bmat
                def B11 (var1 ,var2, var3, var4):
                    return var2
                def B12 (var1 ,var2, var3, var4):
                    return var1
                def B13 (var1 ,var2, var3, var4):
                    return 0
                def B14 (var1 ,var2, var3, var4):
                    return 0

                def B21 (var1 ,var2, var3, var4):
                    return Dc (var1)
                def B22 (var1 ,var2, var3, var4):
                    return var2
                def B23 (var1 ,var2, var3, var4):
                    return 0
                def B24 (var1 ,var2, var3, var4):
                    return 1/(rho_l)

                def B31 (var1 ,var2, var3, var4):
                    return Dc (var1)
                def B32 (var1 ,var2, var3, var4):
                    return 0
                def B33 (var1 ,var2, var3, var4):
                    return var3
                def B34 (var1 ,var2, var3, var4):
                    return 1/rho_g (var4)

                def B41 (var1 ,var2, var3, var4):
                    return (-var2 + var3)*var4/(-1 + var1)
                def B42 (var1 ,var2, var3, var4):
                    return var1*var4/(1 - var1)
                def B43 (var1 ,var2, var3, var4):
                    return var4
                def B44 (var1 ,var2, var3, var4):
                    return var3

                # Cmat
                def Cmat1 (var1, var2, var3, var4):
                    return 0
                def Cmat2 (var1, var2, var3, var4):
                    return  (np.sin(beta)*g*rho_l*var1 + tau_gl (var1, var2, var3, var4)*P_gl (var1)/A - tau_lw (var1, var2)*P_lw (var1)/A + Fbody*var1)/(rho_l*var1)
                def Cmat3 (var1, var2, var3, var4):
                    return -np.sin(beta)*g + (tau_gl (var1, var2, var3, var4)*P_gl (var1)/A + tau_gw (var1, var3, var4)*P_gw (var1)/A + Fbody*(-1 + var1))/(rho_g (var4)*(-1 + var1))
                def Cmat4 (var1, var2, var3, var4):
                    return 0

                # # Dmat
                # def Dmat (var1, var4, Gamma_l, Gamma_g, nu_l, nu_g):
                #     Dmat = [[rho_l*Gamma_l, 0, 0, 0],
                    
                #             [-rho_g (var4)*Gamma_g, 0, 0, 0],
                            
                #             [0, var1*nu_l, 0, 0],
                            
                #             [0, 0, nu_g*(1 - var1), 0]]
                #     return Dmat

            # ===============================================================
            # 3 - QUASILINEAR SYSTEM 
            # U = (alpha_l*rho_l, alpha_g*rho_g, alpha_l*rho_l*u_l, alpha_g*rho_g*u_g)
            # A = I 
            # ===============================================================
            # Matrices
            elif system == "system_3":
                # Amat
                def A11 (var1 ,var2, var3, var4):
                    return 1
                def A12 (var1 ,var2, var3, var4):
                    return 0
                def A13 (var1 ,var2, var3, var4):
                    return 0
                def A14 (var1 ,var2, var3, var4):
                    return 0

                def A21 (var1 ,var2, var3, var4):
                    return 0
                def A22 (var1 ,var2, var3, var4):
                    return 1
                def A23 (var1 ,var2, var3, var4):
                    return 0
                def A24 (var1 ,var2, var3, var4):
                    return 0

                def A31 (var1 ,var2, var3, var4):
                    return 0
                def A32 (var1 ,var2, var3, var4):
                    return 0
                def A33 (var1 ,var2, var3, var4):
                    return 1
                def A34 (var1 ,var2, var3, var4):
                    return 0

                def A41 (var1 ,var2, var3, var4):
                    return 0
                def A42 (var1 ,var2, var3, var4):
                    return 0
                def A43 (var1 ,var2, var3, var4):
                    return 0
                def A44 (var1 ,var2, var3, var4):
                    return 1

                # Bmat
                def B11 (var1 ,var2, var3, var4):
                    return 0
                def B12 (var1 ,var2, var3, var4):
                    return 0
                def B13 (var1 ,var2, var3, var4):
                    return 1
                def B14 (var1 ,var2, var3, var4):
                    return 0

                def B21 (var1 ,var2, var3, var4):
                    return 0
                def B22 (var1 ,var2, var3, var4):
                    return 0
                def B23 (var1 ,var2, var3, var4):
                    return 0
                def B24 (var1 ,var2, var3, var4):
                    return 1

                def B31 (var1 ,var2, var3, var4):
                    return Dc (var1/rho_l)/rho_l + (var1*var2)/(p_factor*rho_l*(rho_l - var1)*(1 - var1/rho_l)) - var3**2/var1**2
                def B32 (var1 ,var2, var3, var4):
                    return var1/(rho_l*(p_factor - (p_factor*var1)/rho_l))
                def B33 (var1 ,var2, var3, var4):
                    return 2*var3/var1
                def B34 (var1 ,var2, var3, var4):
                    return 0

                def B41 (var1 ,var2, var3, var4):
                    return ((var2)/(p_factor*(1 - var1/rho_l)) + (Dc (var1/rho_l)*(p_factor - (p_factor*var1)/rho_l)*var2)/(p_factor*(1 - var1/rho_l)))/rho_l
                def B42 (var1 ,var2, var3, var4):
                    return 1/p_factor - var4**2/var2**2
                def B43 (var1 ,var2, var3, var4):
                    return 0
                def B44 (var1 ,var2, var3, var4):
                    return 2*var4/var2

                # Cmat
                def Cmat1 (var1, var2, var3, var4):
                    return 0
                def Cmat2 (var1, var2, var3, var4):
                    return 0
                def Cmat3 (var1, var2, var3, var4):
                    return g*np.sin (beta)*var1 + tau_lw (var1, var2)*P_lw (var1)/A + tau_gl (var1, var2, var3, var4)*P_gl (var1)/A
                def Cmat4 (var1, var2, var3, var4):
                    return -(np.sin (beta)*g*(var1/rho_l - 1)*var2)/(1 - var1/rho_l) + tau_gw (var1, var3, var4)*P_gw (var1)/A - tau_gl (var1, var2, var3, var4)*P_gl (var1)/A

                # # # Dmat
                # # def Dmat (var1, var4, Gamma_l, Gamma_g, nu_l, nu_g):
                # #     Dmat = [[rho_l*Gamma_l, 0, 0, 0],
                    
                # #             [-rho_g (var4)*Gamma_g, 0, 0, 0],
                            
                # #             [0, var1*nu_l, 0, 0],
                            
                # #             [0, 0, nu_g*(1 - var1), 0]]
                # #     return Dmat


            ############
            
            # Parameters of initial conditions
            if var1_method == "var1_input":
                var1_ref = var1_in
            elif var1_method == "correlation":
                var4= var4_0
                var1_ref = brenth (equilibrium1, lima, limb)
                del var4
            var2_ref = j_l/var1_ref
            var3_ref = j_g/(1 - var1_ref)
            var4_ref = var4_0

            ref = np.array ([var1_ref, var2_ref, var3_ref, var4_ref])
            # print (ref)
            # ===============================================================
            # FINITE ELEMENT SPACES
            # ===============================================================
            # Define constants
            Length = Constant (l)
            mesh = IntervalMesh (nxvector[0], 0, Length)
            h = CellDiameter (mesh)
            coordinates = mesh.coordinates ()
            deltax = l/nxvector[0]          

            if elementspace == "CG_elements":
                e1 = 'Lagrange'
                e2 = 'Lagrange'
                e3 = 'Lagrange'
                e4 = 'Lagrange'
                deg1 = p
                deg2 = p
                deg3 = p
                deg4 = p
            elif elementspace == "TaylorHood":
                e1 = 'Lagrange'
                e2 = 'Lagrange'
                e3 = 'Lagrange'
                e4 = 'Lagrange'
                deg1 = p - 1
                deg2 = p
                deg3 = p
                deg4 = p - 1
            elif elementspace == "Mini":
                e1 = 'Lagrange'
                e2 = 'Lagrange'
                e3 = 'Lagrange'
                e4 = 'Lagrange'
                e2_b = 'Bubble'
                e3_b = 'Bubble'
                deg1 = 1 
                deg2 = 1
                deg3 = 1
                deg4 = 1
                deg2_b = mesh.topology().dim() + 1
                deg3_b = mesh.topology().dim() + 1
            elif elementspace == "IPDG":
                pass

            if any ([elementspace == "CG_elements", elementspace == "TaylorHood"]):
                # Define function space
                element1 = FiniteElement (e1, mesh.ufl_cell (), degree = deg1)
                element2 = FiniteElement (e2, mesh.ufl_cell (), degree = deg2)
                element3 = FiniteElement (e3, mesh.ufl_cell (), degree = deg3)
                element4 = FiniteElement (e4, mesh.ufl_cell (), degree = deg4)
                element = MixedElement ([element1, element2, element3, element4])

                element = MixedElement ([element1, element2, element3, element4])
            elif elementspace == "Mini":
                # Define function space
                element1 = FiniteElement (e1, mesh.ufl_cell (), degree = deg1)
                element2 = FiniteElement (e2, mesh.ufl_cell (), degree = deg2)
                element3 = FiniteElement (e3, mesh.ufl_cell (), degree = deg3)
                element4 = FiniteElement (e4, mesh.ufl_cell (), degree = deg4)

                element2_b = FiniteElement (e2_b, mesh.ufl_cell (), degree = deg2_b)
                element3_b = FiniteElement (e3_b, mesh.ufl_cell (), degree = deg3_b)

                element = MixedElement ([element1, NodalEnrichedElement (element2, element2_b), NodalEnrichedElement (element3, element3_b), element4])

            if dirichlet_type == "nonperiodic":
                V = FunctionSpace (mesh, element)
            elif dirichlet_type == "periodic":
                pbc = PeriodicBoundary()
                V = FunctionSpace (mesh, element, constrained_domain = pbc)    

            # Define variational problem
            dvar = TrialFunction (V)
            v1, v2, v3, v4 = TestFunctions (V)

            # Define functions
            var = Function (V) # solution from (n+1) current step
            var_n = Function (V) # solution from (n) previous converged step
            var_n2 = Function (V) # solution from (n) previous converged step
            var_past = Function (V) # solution from (n-1) previous converged step
            var_ic = Function (V) # reference solution from initial conditions

            # Split mixed functions
            dvar1, dvar2, dvar3, dvar4 = split (dvar)
            var1, var2, var3, var4 = split (var)
            var1_n, var2_n, var3_n, var4_n = split (var_n)
            var1_n2, var2_n2, var3_n2, var4_n2 = split (var_n2)
            var1_past, var2_past, var3_past, var4_past = split (var_past)
            var1_ic, var2_ic, var3_ic, var4_ic = split (var_ic)

            var1_dirichlet = Constant(ref[0])
            var2_dirichlet = Constant(ref[1])
            var3_dirichlet = Constant(ref[2])
            var4_dirichlet = Constant(ref[3])

            bc1 = DirichletBC (V.sub (0), var1_dirichlet, left, "pointwise")
            bc2 = DirichletBC (V.sub (1), var2_dirichlet, left, "pointwise")
            bc3 = DirichletBC (V.sub (2), var3_dirichlet, left, "pointwise")
            bc4 = DirichletBC (V.sub (3), var4_dirichlet, right, "pointwise")
            bcs = [bc1, bc2, bc3, bc4]

            var_init = InitialConditions_SS ()

            # var_n.interpolate (var_init)
            # var_ic.interpolate (var_init)

            # if simulation == 'linear_stability':
            var_ic.interpolate (var_init)
            if time_method == 'BDF1':
                var_n.interpolate (var_init)
            elif time_method == 'CN':
                var_n.interpolate (var_init)
            elif time_method == 'BDF2':
                var_past.interpolate (var_init)

            # ===============================================================
            # TRANSIENT SIMULATIONS
            # ===============================================================
            # Define time discretization properties
            if base_CFL == 'liquid':
                u_CFL = ref[1]
            elif base_CFL == 'gas':
                u_CFL = ref[2]
            elif base_CFL == 'max_velocity':
                u_CFL = max (ref[1], ref[2])

            # Time step
            if timestep == "timestep_input":
                dt = dt_input
            elif timestep == "cfl_based":
                dt = CFL*deltax/(u_CFL)  # Van Zwieten & Smith /lambda_max Holmas (maximum modulus of the eigenvalues of the Jacobian of (4.1) at time 

            if T_simulation == "T_input":     # OPTIONS: T_cfl_based, T_input
                T = T_in
                num_steps = round (T/dt)
            elif T_simulation == "T_cfl_based":
                T = num_steps*dt

            # print("INFO: T     = ", T)
            # print("INFO: dt    = ", dt)
            # print("INFO: steps = ", num_steps)

            # ===============================================================
            # VARIATIONAL FORM
            # ===============================================================
            # Cmat coefficient
            C11_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[0][0])
            C12_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[0][1])
            C13_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[0][2])
            C14_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[0][3])

            C21_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[1][0])
            C22_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[1][1])
            C23_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[1][2])
            C24_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[1][3])

            C31_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[2][0])
            C32_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[2][1])
            C33_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[2][2])
            C34_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[2][3])

            C41_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[3][0])
            C42_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[3][1])
            C43_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[3][2])
            C44_ref = Constant (Cmat_lin (ref[0], ref[1], ref[2], ref[3])[3][3])

            # Cmat_lin
            def C11 (var1 ,var2, var3, var4):
                return C11_ref*var1
            def C12 (var1 ,var2, var3, var4):
                return C12_ref*var2
            def C13 (var1 ,var2, var3, var4):
                return C13_ref*var3
            def C14 (var1 ,var2, var3, var4):
                return C14_ref*var4

            def C21 (var1 ,var2, var3, var4):
                return C21_ref*var1
            def C22 (var1 ,var2, var3, var4):
                return C22_ref*var2
            def C23 (var1 ,var2, var3, var4):
                return C23_ref*var3
            def C24 (var1 ,var2, var3, var4):
                return C24_ref*var4

            def C31 (var1 ,var2, var3, var4):
                return C31_ref*var1
            def C32 (var1 ,var2, var3, var4):
                return C32_ref*var2
            def C33 (var1 ,var2, var3, var4):
                return C33_ref*var3
            def C34 (var1 ,var2, var3, var4):
                return C34_ref*var4

            def C41 (var1 ,var2, var3, var4):
                return C41_ref*var1
            def C42 (var1 ,var2, var3, var4):
                return C42_ref*var2
            def C43 (var1 ,var2, var3, var4):
                return C43_ref*var3
            def C44 (var1 ,var2, var3, var4):
                return C44_ref*var4

            if simulation == 'linear_stability':
                # Bmat
                Bm1 = ( inner (B11 (ref[0], ref[1], ref[2], ref[3])*Dx (var1, 0), v1) \
                        + inner (B12 (ref[0], ref[1], ref[2], ref[3])*Dx (var2, 0), v1) \
                        + inner (B13 (ref[0], ref[1], ref[2], ref[3])*Dx (var3, 0), v1) \
                        + inner (B14 (ref[0], ref[1], ref[2], ref[3])*Dx (var4, 0), v1) )*dx 

                Bm2 = ( inner (B21 (ref[0], ref[1], ref[2], ref[3])*Dx (var1, 0), v2) \
                        + inner (B22 (ref[0], ref[1], ref[2], ref[3])*Dx (var2, 0), v2) \
                        + inner (B23 (ref[0], ref[1], ref[2], ref[3])*Dx (var3, 0), v2) \
                        + inner (B24 (ref[0], ref[1], ref[2], ref[3])*Dx (var4, 0), v2) )*dx 

                Bm3 = ( inner (B31 (ref[0], ref[1], ref[2], ref[3])*Dx (var1, 0), v3) \
                        + inner (B32 (ref[0], ref[1], ref[2], ref[3])*Dx (var2, 0), v3) \
                        + inner (B33 (ref[0], ref[1], ref[2], ref[3])*Dx (var3, 0), v3) \
                        + inner (B34 (ref[0], ref[1], ref[2], ref[3])*Dx (var4, 0), v3) )*dx 

                Bm4 = ( inner (B41 (ref[0], ref[1], ref[2], ref[3])*Dx (var1, 0), v4) \
                        + inner (B42 (ref[0], ref[1], ref[2], ref[3])*Dx (var2, 0), v4) \
                        + inner (B43 (ref[0], ref[1], ref[2], ref[3])*Dx (var3, 0), v4) \
                        + inner (B44 (ref[0], ref[1], ref[2], ref[3])*Dx (var4, 0), v4) )*dx
                
                # Cmat_lin
                Cm1_lin = ( inner (C11 (var1, var2, var3, var4), v1) \
                + inner (C12 (var1, var2, var3, var4), v1) \
                + inner (C13 (var1, var2, var3, var4), v1) \
                + inner (C14 (var1, var2, var3, var4), v1) )*dx

                Cm2_lin = ( inner (C21 (var1, var2, var3, var4), v2) \
                + inner (C22 (var1, var2, var3, var4), v2) \
                + inner (C23 (var1, var2, var3, var4), v2) \
                + inner (C24 (var1, var2, var3, var4), v2) )*dx

                Cm3_lin = ( inner (C31 (var1, var2, var3, var4),v3) \
                + inner (C32 (var1, var2, var3, var4),v3) \
                + inner (C33 (var1, var2, var3, var4),v3) \
                + inner (C34 (var1, var2, var3, var4),v3) )*dx

                Cm4_lin = ( inner (C41 (var1, var2, var3, var4), v4) \
                + inner (C42 (var1, var2, var3, var4), v4) \
                + inner (C43 (var1, var2, var3, var4), v4) \
                + inner (C44 (var1, var2, var3, var4), v4) )*dx

                if discretization == "fully_discretized":
                    if any ([time_method == 'BDF1', time_method == 'CN']):
                        # Bmat_n
                        Bm1_n = ( inner (B11 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_n, 0), v1) \
                                + inner (B12 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_n, 0), v1) \
                                + inner (B13 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_n, 0), v1) \
                                + inner (B14 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_n, 0), v1) )*dx 

                        Bm2_n = ( inner (B21 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_n, 0), v2) \
                                + inner (B22 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_n, 0), v2) \
                                + inner (B23 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_n, 0), v2) \
                                + inner (B24 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_n, 0), v2) )*dx 

                        Bm3_n = ( inner (B31 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_n, 0), v3) \
                                + inner (B32 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_n, 0), v3) \
                                + inner (B33 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_n, 0), v3) \
                                + inner (B34 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_n, 0), v3) )*dx 

                        Bm4_n = ( inner (B41 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_n, 0), v4) \
                                + inner (B42 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_n, 0), v4) \
                                + inner (B43 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_n, 0), v4) \
                                + inner (B44 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_n, 0), v4) )*dx
                        
                        # Cmat_lin_n
                        Cm1_lin_n = ( inner (C11 (var1_n, var2_n, var3_n, var4_n),v1) \
                        + inner (C12 (var1_n, var2_n, var3_n, var4_n),v1) \
                        + inner (C13 (var1_n, var2_n, var3_n, var4_n),v1) \
                        + inner (C14 (var1_n, var2_n, var3_n, var4_n),v1) )*dx

                        Cm2_lin_n = ( inner (C21 (var1_n, var2_n, var3_n, var4_n), v2) \
                        + inner (C22 (var1_n, var2_n, var3_n, var4_n), v2) \
                        + inner (C23 (var1_n, var2_n, var3_n, var4_n), v2) \
                        + inner (C24 (var1_n, var2_n, var3_n, var4_n), v2) )*dx

                        Cm3_lin_n = ( inner (C31 (var1_n, var2_n, var3_n, var4_n),v3) \
                        + inner (C32 (var1_n, var2_n, var3_n, var4_n),v3) \
                        + inner (C33 (var1_n, var2_n, var3_n, var4_n),v3) \
                        + inner (C34 (var1_n, var2_n, var3_n, var4_n),v3) )*dx
                        
                        Cm4_lin_n = ( inner (C41 (var1_n, var2_n, var3_n, var4_n),v4) \
                        + inner (C42 (var1_n, var2_n, var3_n, var4_n),v4) \
                        + inner (C43 (var1_n, var2_n, var3_n, var4_n),v4) \
                        + inner (C44 (var1_n, var2_n, var3_n, var4_n),v4) )*dx

                    elif time_method == 'BDF2':
                        # Bmat_n
                        Bm1_n = ( inner (B11 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_n, 0), v1) \
                                + inner (B12 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_n, 0), v1) \
                                + inner (B13 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_n, 0), v1) \
                                + inner (B14 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_n, 0), v1) )*dx 

                        Bm2_n = ( inner (B21 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_n, 0), v2) \
                                + inner (B22 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_n, 0), v2) \
                                + inner (B23 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_n, 0), v2) \
                                + inner (B24 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_n, 0), v2) )*dx 

                        Bm3_n = ( inner (B31 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_n, 0), v3) \
                                + inner (B32 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_n, 0), v3) \
                                + inner (B33 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_n, 0), v3) \
                                + inner (B34 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_n, 0), v3) )*dx 

                        Bm4_n = ( inner (B41 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_n, 0), v4) \
                                + inner (B42 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_n, 0), v4) \
                                + inner (B43 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_n, 0), v4) \
                                + inner (B44 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_n, 0), v4) )*dx
                        
                        # Cmat_lin_n
                        Cm1_lin_n = ( inner (C11 (var1_n, var2_n, var3_n, var4_n),v1) \
                        + inner (C12 (var1_n, var2_n, var3_n, var4_n),v1) \
                        + inner (C13 (var1_n, var2_n, var3_n, var4_n),v1) \
                        + inner (C14 (var1_n, var2_n, var3_n, var4_n),v1) )*dx

                        Cm2_lin_n = ( inner (C21 (var1_n, var2_n, var3_n, var4_n), v2) \
                        + inner (C22 (var1_n, var2_n, var3_n, var4_n), v2) \
                        + inner (C23 (var1_n, var2_n, var3_n, var4_n), v2) \
                        + inner (C24 (var1_n, var2_n, var3_n, var4_n), v2) )*dx

                        Cm3_lin_n = ( inner (C31 (var1_n, var2_n, var3_n, var4_n),v3) \
                        + inner (C32 (var1_n, var2_n, var3_n, var4_n),v3) \
                        + inner (C33 (var1_n, var2_n, var3_n, var4_n),v3) \
                        + inner (C34 (var1_n, var2_n, var3_n, var4_n),v3) )*dx
                        
                        Cm4_lin_n = ( inner (C41 (var1_n, var2_n, var3_n, var4_n),v4) \
                        + inner (C42 (var1_n, var2_n, var3_n, var4_n),v4) \
                        + inner (C43 (var1_n, var2_n, var3_n, var4_n),v4) \
                        + inner (C44 (var1_n, var2_n, var3_n, var4_n),v4) )*dx

                        # Bmat_past
                        Bm1_past = ( inner (B11 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_past, 0), v1) \
                                + inner (B12 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_past, 0), v1) \
                                + inner (B13 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_past, 0), v1) \
                                + inner (B14 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_past, 0), v1) )*dx 

                        Bm2_past = ( inner (B21 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_past, 0), v2) \
                                + inner (B22 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_past, 0), v2) \
                                + inner (B23 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_past, 0), v2) \
                                + inner (B24 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_past, 0), v2) )*dx 

                        Bm3_past = ( inner (B31 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_past, 0), v3) \
                                + inner (B32 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_past, 0), v3) \
                                + inner (B33 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_past, 0), v3) \
                                + inner (B34 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_past, 0), v3) )*dx 

                        Bm4_past = ( inner (B41 (ref[0], ref[1], ref[2], ref[3])*Dx (var1_past, 0), v4) \
                                + inner (B42 (ref[0], ref[1], ref[2], ref[3])*Dx (var2_past, 0), v4) \
                                + inner (B43 (ref[0], ref[1], ref[2], ref[3])*Dx (var3_past, 0), v4) \
                                + inner (B44 (ref[0], ref[1], ref[2], ref[3])*Dx (var4_past, 0), v4) )*dx
                        
                        # Cmat_lin_past
                        Cm1_lin_past = ( inner (C11 (var1_past, var2_past, var3_past, var4_past), v1) \
                        + inner (C12 (var1_past, var2_past, var3_past, var4_past), v1) \
                        + inner (C13 (var1_past, var2_past, var3_past, var4_past), v1) \
                        + inner (C14 (var1_past, var2_past, var3_past, var4_past), v1) )*dx

                        Cm2_lin_past = ( inner (C21 (var1_past, var2_past, var3_past, var4_past), v2) \
                        + inner (C22 (var1_past, var2_past, var3_past, var4_past), v2) \
                        + inner (C23 (var1_past, var2_past, var3_past, var4_past), v2) \
                        + inner (C24 (var1_past, var2_past, var3_past, var4_past), v2) )*dx

                        Cm3_lin_past = ( inner (C31 (var1_past, var2_past, var3_past, var4_past),v3) \
                        + inner (C32 (var1_past, var2_past, var3_past, var4_past),v3) \
                        + inner (C33 (var1_past, var2_past, var3_past, var4_past),v3) \
                        + inner (C34 (var1_past, var2_past, var3_past, var4_past),v3) )*dx

                        Cm4_lin_past = ( inner (C41 (var1_past, var2_past, var3_past, var4_past), v4) \
                        + inner (C42 (var1_past, var2_past, var3_past, var4_past), v4) \
                        + inner (C43 (var1_past, var2_past, var3_past, var4_past), v4) \
                        + inner (C44 (var1_past, var2_past, var3_past, var4_past), v4) )*dx

            # ===============================================================
            # FULLY-DISCRETIZED
            # ===============================================================
            # L2 norm for each time step
            L2norm_var1 = []
            L2norm_var2 = []
            L2norm_var3 = []
            L2norm_var4 = []
            # Time vector for L2 norm computation
            timevector = np.linspace(0, T, num_steps)
            # print("control")

            # Step in time
            t = 0
    # ===============================================================
    # CURRENT TIME BDF1
    # ===============================================================
            if time_method == 'BDF1':
                # Method
                # print('INFO: first-order Backward differentiation formula time integration.')

                # Initial conditions
                (var1_n, var2_n, var3_n, var4_n) = var_n.split (deepcopy = True)
                # Compute  nodal values (initial)
                n_val1_n = np.array (var1_n.vector ())
                n_val2_n = np.array (var2_n.vector ())
                n_val3_n = np.array (var3_n.vector ())
                n_val4_n = np.array (var4_n.vector ())

                nodal_var1_n = n_val1_n [::-1]
                nodal_var2_n = n_val2_n [::-1]
                nodal_var3_n = n_val3_n [::-1]
                nodal_var4_n = n_val4_n [::-1]

                # print("n", nodal_var1_n)

                nodes_var1_n = len (nodal_var1_n)
                nodes_var2_n = len (nodal_var2_n)
                nodes_var3_n = len (nodal_var3_n)
                nodes_var4_n = len (nodal_var4_n)

                # print ("dof subspace var1 =", nodes_var1_n)
                # print ("dof subspace var2 =", nodes_var2_n)
                # print ("dof subspace var3 =", nodes_var3_n)
                # print ("dof subspace var4 =", nodes_var4_n)

                # Compute vertex values (initial step)
                v_var1_n = var1_n.compute_vertex_values (mesh)
                v_var2_n = var2_n.compute_vertex_values (mesh)
                v_var3_n = var3_n.compute_vertex_values (mesh)
                v_var4_n = var4_n.compute_vertex_values (mesh)

                vert_var1_n = []
                vert_var2_n = []
                vert_var3_n = []
                vert_var4_n = []

                for cond_vert in range (len(coordinates)):
                    vert_var1_n.append (v_var1_n[cond_vert])
                    vert_var2_n.append (v_var2_n[cond_vert])
                    vert_var3_n.append (v_var3_n[cond_vert])
                    vert_var4_n.append (v_var4_n[cond_vert])
            
                vert_var1_n = np.asarray (vert_var1_n)
                vert_var2_n = np.asarray (vert_var2_n)
                vert_var3_n = np.asarray (vert_var3_n)
                vert_var4_n = np.asarray (vert_var4_n)

                # print("var1_n = ", vert_var1_n)

                # # Well-posedness check
                # i = 0
                # for var1_test, var2_test, var3_test, var4_test in zip (vert_var1_n, vert_var2_n, vert_var3_n, vert_var4_n):
                #     Aeval = np.asarray (Amat (var1_test, var2_test, var3_test, var4_test))
                #     Beval = np.asarray (Bmat (var1_test, var2_test, var3_test, var4_test))

                #     m_test, _ = eig (Beval, b = Aeval, overwrite_a = True, overwrite_b = True, check_finite = True)

                #     listreal = []
                #     listimag = []

                #     for cond0 in range (len (m_test)):
                #         realpart = m_test [cond0].real                
                #         imagpart = m_test [cond0].imag
                        
                #         listimag.append (imagpart)
                #         listreal.append (realpart)
                    
                #     if any ([ cond1 != 0 for cond1 in listimag ]) :
                #         pass#print("INFO: ill-posed equation system")
                #         break
                #     else:
                #         i += 1
                #     break

                # # Condition for low liquid level of transition to intermittent flow
                # if any ([cond2 <= 0 for cond2 in vert_var1_n]) :
                #     # print("INFO: gas phase dominant. Low liquid level. Liquid equations vanish")
                #     break
                # elif any ([cond2 >= 1 for cond2 in vert_var1_n]):
                #     # print("INFO: liquid phase dominant. Intermittent flow. Gas equations vanish")
                #     break
                # else:
                #     pass
                # # Condition for positive velocities and pressure
                # if any ([ cond3 <= 0 for cond3 in vert_var2_n]) :
                #     # print("INFO: negative liquid velocity")
                #     break
                # else:
                #     pass

                # if any ([ cond4 <= 0 for cond4 in vert_var3_n]) :
                #     # print("INFO: negative gas velocity")
                #     break
                # else:
                #     pass

                # if any ([ cond5 <= 0 for cond5 in vert_var4_n]) :
                #     # print("INFO: negative pressure")
                #     break
                # else:
                #     pass
                
                # # Reynolds check
                # Rel_n = Re_l (vert_var1_n, vert_var2_n)
                # Reg_n = Re_g (vert_var1_n, vert_var3_n, vert_var4_n)

                # if any ([ cond3 <= 1180 for cond3 in Rel_n]):
                # # Rel < 1180:
                #     pass #print("Liquid laminar")
                # elif  any ([ cond3 > 1180 for cond3 in Rel_n]):
                #     pass

                # if any ([ cond4 <= 1180 for cond4 in Reg_n]):
                #     pass #print("Gas laminar")
                # elif  any ([ cond4 > 1180 for cond4 in Reg_n]):
                #     pass

                # # Plot solution
                # plt.figure (5, figsize = (8, 4))
                # plt.xlim (0, l)
                # plt.ylim (0, 1)
                # plt.grid (True, which = "both")
                # plot (var1_n, title = r"$\alpha_l$", rescale = True)

                # plt.figure (6, figsize = (8, 4))
                # plt.xlim (0, l)
                # plt.ylim (min(vert_var2_n), max(vert_var2_n))
                # plt.grid (True, which = "both")

                # plot (var2_n, title = r"$u_l$", rescale = True)

                # plt.figure (7, figsize = (8, 4))
                # plt.xlim (0, l)
                # plt.grid (True, which = "both")
                # plot (var3_n, title = r"$u_g$", rescale = True)

                # plt.figure (8, figsize = (8, 4))
                # plt.xlim (0, l)
                # plt.grid (True, which = "both")
                # plot (var4_n, title = r"$p_i$", rescale = True)
                
                # Save solution
                # ff_var1 << var1_n
                # ff_var2 << var2_n
                # ff_var3 << var3_n
                # ff_var4 << var4_n

                # Time stepping
                t = dt
                
                # Print progress
                # clear_output ()
                # print ("Iteration :", int (round (t/dt)), "of", num_steps)
                # print ("Time      :", t, "s")
                                
                # Parameters
                a0 = 1
                a1 = -1
                a2 = 0
                theta = 1
                
                # Amat
                Am1 = ( (a0*var1 + a1*var1_n + a2*var1_past)/dt*v1 )*dx 
                Am2 = ( (a0*var2 + a1*var2_n + a2*var1_past)/dt*v2 )*dx 
                Am3 = ( (a0*var3 + a1*var3_n + a2*var1_past)/dt*v3 )*dx 
                Am4 = ( (a0*var4 + a1*var4_n + a2*var1_past)/dt*v4 )*dx

                # Variational form
                F = visc*(Cm1_lin + Cm2_lin + Cm3_lin + Cm4_lin) - (Bm1 + Bm2 + Bm3 + Bm4)
                F_n = visc*(Cm1_lin_n + Cm2_lin_n + Cm3_lin_n + Cm4_lin_n) - (Bm1_n + Bm2_n + Bm3_n + Bm4_n)
                R = (Am1 + Am2 + Am3 + Am4) - theta*F - (1 - theta)*F_n 

                # Compute directional derivative about u in the direction of du (Jacobian)
                dF = derivative (R, var, dvar)

                problem = NonlinearVariationalProblem (R, var, bcs = bcs, J = dF, form_compiler_parameters = ffc_options)
                solver  = NonlinearVariationalSolver (problem)
                prm     = solver.parameters
                # info (prm, True)
                
                # Nonlinear solver
                prm ['nonlinear_solver'] = 'newton'
                # prm ['print_matrix']     = False #True
                # prm ['print_rhs']        = False #True
                # prm ['symmetric']        = False #True
                
                # Linear solver
                # prm ['newton_solver']['absolute_tolerance']      = 1e-1 #1E-8
                # prm ['newton_solver']['convergence_criterion']   = 'residual' #'residual' 'incremental'
                # prm ['newton_solver']['error_on_nonconvergence'] = True
                prm ['newton_solver']['linear_solver']           = 'umfpack' # 'bicgstab' 'cg' 'gmres' 'minres' 'petsc' 'richardson' 'superlu_dist' 'tfqmr' 'umfpack'
                # prm ['newton_solver']['maximum_iterations']      = 10000
                # prm ['newton_solver']['preconditioner']          = 'ilu' # 'ilu' 'icc' 'petsc_amg' 'sor'
                # prm ['newton_solver']['relative_tolerance']      = 1e-1
                # prm ['newton_solver']['relaxation_parameter']    = 1.0
                # prm ['newton_solver']['report']                  = True

                # Krylov solver
                # prm ['newton_solver']['krylov_solver']['absolute_tolerance']       = 1e-3 #1E-9
                # #     prm ['newton_solver']['krylov_solver']['error_on_nonconvergence']  = True
                # prm ['newton_solver']['krylov_solver']['maximum_iterations']       = 10000 # 500000
                # prm ['newton_solver']['krylov_solver']["monitor_convergence"]      = True
                # prm ['newton_solver']['krylov_solver']["nonzero_initial_guess"]    = True #False
                # prm ['newton_solver']['krylov_solver']['relative_tolerance']       = 1e-3
                # prm ['newton_solver']['krylov_solver']['report']                   = True
                    
                # LU solver
                # prm ['newton_solver']['lu_solver']['report']    = True
                # prm ['newton_solver']['lu_solver']['symmetric'] = False
                # prm ['newton_solver']['lu_solver']['verbose']   = True

                solver.solve ()  

                # Split of the vector var
                (var1, var2, var3, var4) = var.split (deepcopy = True)

                # L2 norm
                # infonorm = var1.vector().norm("l2")
                # print("Solution vector norm (0): {!r}".format(infonorm)) 
                L2norm_var1.append (var1.vector().norm("l2"))
                L2norm_var2.append (var2.vector().norm("l2"))   
                L2norm_var3.append (var3.vector().norm("l2"))   
                L2norm_var4.append (var4.vector().norm("l2"))   

                # Nodal values (current)
                n_val1_n1 = np.array(var1.vector())
                n_val2_n1 = np.array(var2.vector())
                n_val3_n1 = np.array(var3.vector())
                n_val4_n1 = np.array(var4.vector())

                nodal_var1_n1 = n_val1_n1 [::-1]
                nodal_var2_n1 = n_val2_n1 [::-1]
                nodal_var3_n1 = n_val3_n1 [::-1]
                nodal_var4_n1 = n_val4_n1 [::-1]

                # print("n+1", nodal_var1_n1)

                # Compute vertex values (next step)
                vert_var1 = var1.compute_vertex_values (mesh)
                vert_var2 = var2.compute_vertex_values (mesh)
                vert_var3 = var3.compute_vertex_values (mesh)
                vert_var4 = var4.compute_vertex_values (mesh)

                vert_var1_n1 = []
                vert_var2_n1 = []
                vert_var3_n1 = []
                vert_var4_n1 = []

                for cond_vert in range (len(coordinates)):
                    vert_var1_n1.append (vert_var1[cond_vert])
                    vert_var2_n1.append (vert_var2[cond_vert])
                    vert_var3_n1.append (vert_var3[cond_vert])
                    vert_var4_n1.append (vert_var4[cond_vert])
            
                vert_var1_n1 = np.asarray (vert_var1_n1)
                vert_var2_n1 = np.asarray (vert_var2_n1)
                vert_var3_n1 = np.asarray (vert_var3_n1)
                vert_var4_n1 = np.asarray (vert_var4_n1)

                # print ("var1_n1 = ", vert_var1_n1)

                # # Well-posedness check
                # i = 0
                # for var1_test, var2_test, var3_test, var4_test in zip (vert_var1_n1, vert_var2_n1, vert_var3_n1, vert_var4_n1):
                #     Aeval = np.asarray (Amat (var1_test, var2_test, var3_test, var4_test))
                #     Beval = np.asarray (Bmat (var1_test, var2_test, var3_test, var4_test))

                #     m_test, _ = eig (Beval, b = Aeval, overwrite_a = True, overwrite_b = True, check_finite = True)

                #     listreal = []
                #     listimag = []

                #     for cond0 in range (len (m_test)):
                #         realpart = m_test [cond0].real                
                #         imagpart = m_test [cond0].imag
                        
                #         listimag.append (imagpart)
                #         listreal.append (realpart)
                    
                #     if any ([ cond1 != 0 for cond1 in listimag ]) :
                #         # print("INFO: ill-posed equation system")
                #         break
                #     else:
                #         i += 1
                #     break

                # # Condition for low liquid level of transition to intermittent flow
                # if any ([cond2 <= 0 for cond2 in vert_var1_n1]) :
                #     # print("INFO: gas phase dominant. Low liquid level. Liquid equations vanish")
                #     break
                # elif any ([cond2 >= 1 for cond2 in vert_var1_n1]):
                #     # print("INFO: liquid phase dominant. Intermittent flow. Gas equations vanish")
                #     break
                # else:
                #     pass
                # # Condition for positive velocities and pressure
                # if any ([ cond3 <= 0 for cond3 in vert_var2_n1]) :
                #     # print("INFO: negative liquid velocity")
                #     break
                # else:
                #     pass

                # if any ([ cond4 <= 0 for cond4 in vert_var3_n1]) :
                #     # print("INFO: negative gas velocity")
                #     break
                # else:
                #     pass

                # if any ([ cond5 <= 0 for cond5 in vert_var4_n1]) :
                #     # print("INFO: negative pressure")
                #     break
                # else:
                #     pass
                
                # # Reynolds check
                # Rel_n1 = Re_l (vert_var1_n1, vert_var2_n1)
                # Reg_n1 = Re_g (vert_var1_n1, vert_var3_n1, vert_var4_n1)

                # if any ([ cond3 <= 1180 for cond3 in Rel_n1]):
                # # Rel < 1180:
                #     pass #print("Liquid laminar")
                # elif  any ([ cond3 > 1180 for cond3 in Rel_n1]):
                #     pass

                # if any ([ cond4 <= 1180 for cond4 in Reg_n1]):
                #     pass #print("Gas laminar")
                # elif  any ([ cond4 > 1180 for cond4 in Reg_n1]):
                #     pass

# ===============================================================
# CURRENT TIME CN
# ===============================================================
            elif time_method == 'CN':
                # Method
                # print('INFO: Crank-Nicolson/trapezoidal time integration.')
                
                # Initial conditions
                (var1_n, var2_n, var3_n, var4_n) = var_n.split (deepcopy = True)
                # Compute  nodal values (initial)
                n_val1_n = np.array (var1_n.vector ())
                n_val2_n = np.array (var2_n.vector ())
                n_val3_n = np.array (var3_n.vector ())
                n_val4_n = np.array (var4_n.vector ())

                nodal_var1_n = n_val1_n [::-1]
                nodal_var2_n = n_val2_n [::-1]
                nodal_var3_n = n_val3_n [::-1]
                nodal_var4_n = n_val4_n [::-1]

                # print("n", nodal_var1_n)

                nodes_var1_n = len (nodal_var1_n)
                nodes_var2_n = len (nodal_var2_n)
                nodes_var3_n = len (nodal_var3_n)
                nodes_var4_n = len (nodal_var4_n)

                # print ("dof subspace var1 =", nodes_var1_n)
                # print ("dof subspace var2 =", nodes_var2_n)
                # print ("dof subspace var3 =", nodes_var3_n)
                # print ("dof subspace var4 =", nodes_var4_n)

                # Compute vertex values (initial step)
                v_var1_n = var1_n.compute_vertex_values (mesh)
                v_var2_n = var2_n.compute_vertex_values (mesh)
                v_var3_n = var3_n.compute_vertex_values (mesh)
                v_var4_n = var4_n.compute_vertex_values (mesh)

                vert_var1_n = []
                vert_var2_n = []
                vert_var3_n = []
                vert_var4_n = []

                for cond_vert in range (len(coordinates)):
                    vert_var1_n.append (v_var1_n[cond_vert])
                    vert_var2_n.append (v_var2_n[cond_vert])
                    vert_var3_n.append (v_var3_n[cond_vert])
                    vert_var4_n.append (v_var4_n[cond_vert])
            
                vert_var1_n = np.asarray (vert_var1_n)
                vert_var2_n = np.asarray (vert_var2_n)
                vert_var3_n = np.asarray (vert_var3_n)
                vert_var4_n = np.asarray (vert_var4_n)

                # print("var1_n = ", vert_var1_n)

                # # Well-posedness check
                # i = 0
                # for var1_test, var2_test, var3_test, var4_test in zip (vert_var1_n, vert_var2_n, vert_var3_n, vert_var4_n):
                #     Aeval = np.asarray (Amat (var1_test, var2_test, var3_test, var4_test))
                #     Beval = np.asarray (Bmat (var1_test, var2_test, var3_test, var4_test))

                #     m_test, _ = eig (Beval, b = Aeval, overwrite_a = True, overwrite_b = True, check_finite = True)

                #     listreal = []
                #     listimag = []

                #     for cond0 in range (len (m_test)):
                #         realpart = m_test [cond0].real                
                #         imagpart = m_test [cond0].imag
                        
                #         listimag.append (imagpart)
                #         listreal.append (realpart)
                    
                #     if any ([ cond1 != 0 for cond1 in listimag ]) :
                #         print("INFO: ill-posed equation system")
                #         break
                #     else:
                #         i += 1
                #     break

                # # Condition for low liquid level of transition to intermittent flow
                # if any ([cond2 <= 0 for cond2 in vert_var1_n]) :
                #     print("INFO: gas phase dominant. Low liquid level. Liquid equations vanish")
                #     break
                # elif any ([cond2 >= 1 for cond2 in vert_var1_n]):
                #     print("INFO: liquid phase dominant. Intermittent flow. Gas equations vanish")
                #     break
                # else:
                #     pass
                # # Condition for positive velocities and pressure
                # if any ([ cond3 <= 0 for cond3 in vert_var2_n]) :
                #     print("INFO: negative liquid velocity")
                #     break
                # else:
                #     pass

                # if any ([ cond4 <= 0 for cond4 in vert_var3_n]) :
                #     print("INFO: negative gas velocity")
                #     break
                # else:
                #     pass

                # if any ([ cond5 <= 0 for cond5 in vert_var4_n]) :
                #     print("INFO: negative pressure")
                #     break
                # else:
                #     pass
                
                # # Reynolds check
                # Rel_n = Re_l (vert_var1_n, vert_var2_n)
                # Reg_n = Re_g (vert_var1_n, vert_var3_n, vert_var4_n)

                # if any ([ cond3 <= 1180 for cond3 in Rel_n]):
                # # Rel < 1180:
                #     print("Liquid laminar")
                # elif  any ([ cond3 > 1180 for cond3 in Rel_n]):
                #     pass

                # if any ([ cond4 <= 1180 for cond4 in Reg_n]):
                #     print("Gas laminar")
                # elif  any ([ cond4 > 1180 for cond4 in Reg_n]):
                #     pass

                # # c = conditional (ge (var1, 1), 2, info (var1, verbose=False))
                # c = conditional (le (var1, 0), str(INFO: liquid phase dominant, which leads to intermittent flow. Gas equations vanish), 1)

                # # Plot solution
                # plt.figure (5, figsize = (8, 4))
                # plt.xlim (0, l)
                # plt.ylim (0, 1)
                # plt.grid (True, which = "both")
                # plot (var1_n, title = r"$\alpha_l$", rescale = True)

                # plt.figure (6, figsize = (8, 4))
                # plt.xlim (0, l)
                # plt.ylim (min(vert_var2_n), max(vert_var2_n))
                # plt.grid (True, which = "both")

                # plot (var2_n, title = r"$u_l$", rescale = True)

                # plt.figure (7, figsize = (8, 4))
                # plt.xlim (0, l)
                # plt.grid (True, which = "both")
                # plot (var3_n, title = r"$u_g$", rescale = True)

                # plt.figure (8, figsize = (8, 4))
                # plt.xlim (0, l)
                # plt.grid (True, which = "both")
                # plot (var4_n, title = r"$p_i$", rescale = True)
                
                # Save solution
                # ff_var1 << var1_n
                # ff_var2 << var2_n
                # ff_var3 << var3_n
                # ff_var4 << var4_n


                # Time stepping
                t = dt
                
                # Print progress
                # clear_output ()
                # print ("Iteration :", int (round (t/dt)), "of", num_steps)
                # print ("Time      :", t, "s")

                # Parameters
                a0 = 1
                a1 = -1
                a2 = 0
                theta = 1/2

                # Amat
                Am1 = ( (a0*var1 + a1*var1_n + a2*var1_past)/dt*v1 )*dx 
                Am2 = ( (a0*var2 + a1*var2_n + a2*var1_past)/dt*v2 )*dx 
                Am3 = ( (a0*var3 + a1*var3_n + a2*var1_past)/dt*v3 )*dx 
                Am4 = ( (a0*var4 + a1*var4_n + a2*var1_past)/dt*v4 )*dx

                # Variational form
                F = visc*(Cm1_lin + Cm2_lin + Cm3_lin + Cm4_lin) - (Bm1 + Bm2 + Bm3 + Bm4)
                F_n = visc*(Cm1_lin_n + Cm2_lin_n + Cm3_lin_n + Cm4_lin_n) - (Bm1_n + Bm2_n + Bm3_n + Bm4_n)
                R = (Am1 + Am2 + Am3 + Am4) - theta*F - (1 - theta)*F_n 

                # Compute directional derivative about u in the direction of du (Jacobian)
                dF = derivative (R, var, dvar)

                problem = NonlinearVariationalProblem (R, var, bcs = bcs, J = dF, form_compiler_parameters = ffc_options)
                solver  = NonlinearVariationalSolver (problem)
                prm     = solver.parameters
                info (prm, True)
                    
                # Nonlinear solver
                prm ['nonlinear_solver'] = 'newton'
                # prm ['print_matrix']     = False #True
                # prm ['print_rhs']        = False #True
                # prm ['symmetric']        = False #True
                
                # linear solver
                # prm ['newton_solver']['absolute_tolerance']      = 1e-1 #1E-8
                # prm ['newton_solver']['convergence_criterion']   = 'residual' #'residual' 'incremental'
                # prm ['newton_solver']['error_on_nonconvergence'] = True
                prm ['newton_solver']['linear_solver']           = 'umfpack' # 'bicgstab' 'cg' 'gmres' 'minres' 'petsc' 'richardson' 'superlu_dist' 'tfqmr' 'umfpack'
                # prm ['newton_solver']['maximum_iterations']      = 10000
                # prm ['newton_solver']['preconditioner']          = 'ilu' # 'ilu' 'icc' 'petsc_amg' 'sor'
                # prm ['newton_solver']['relative_tolerance']      = 1e-1
                # prm ['newton_solver']['relaxation_parameter']    = 1.0
                # prm ['newton_solver']['report']                  = True

                # Krylov solver
                # prm ['newton_solver']['krylov_solver']['absolute_tolerance']       = 1e-3 #1E-9
                # #     prm ['newton_solver']['krylov_solver']['error_on_nonconvergence']  = True
                # prm ['newton_solver']['krylov_solver']['maximum_iterations']       = 10000 # 500000
                # prm ['newton_solver']['krylov_solver']["monitor_convergence"]      = True
                # prm ['newton_solver']['krylov_solver']["nonzero_initial_guess"]    = True #False
                # prm ['newton_solver']['krylov_solver']['relative_tolerance']       = 1e-3
                # prm ['newton_solver']['krylov_solver']['report']                   = True
                
                # LU solver
                # prm ['newton_solver']['lu_solver']['report']    = True
                # prm ['newton_solver']['lu_solver']['symmetric'] = False
                # prm ['newton_solver']['lu_solver']['verbose']   = True

                solver.solve ()  

                # Split of the vector var
                (var1, var2, var3, var4) = var.split (deepcopy = True)

                # L2 norm
                # infonorm = var1.vector().norm("l2")
                # print("Solution vector norm (0): {!r}".format(infonorm)) 
                # L2norm_var1.append (var1.vector().norm("l2"))
                # L2norm_var2.append (var2.vector().norm("l2"))   
                # L2norm_var3.append (var3.vector().norm("l2"))   
                # L2norm_var4.append (var4.vector().norm("l2"))   

                # Nodal values (current)
                n_val1_n1 = np.array(var1.vector())
                n_val2_n1 = np.array(var2.vector())
                n_val3_n1 = np.array(var3.vector())
                n_val4_n1 = np.array(var4.vector())

                nodal_var1_n1 = n_val1_n1 [::-1]
                nodal_var2_n1 = n_val2_n1 [::-1]
                nodal_var3_n1 = n_val3_n1 [::-1]
                nodal_var4_n1 = n_val4_n1 [::-1]

                # Compute vertex values (next step)
                vert_var1 = var1.compute_vertex_values (mesh)
                vert_var2 = var2.compute_vertex_values (mesh)
                vert_var3 = var3.compute_vertex_values (mesh)
                vert_var4 = var4.compute_vertex_values (mesh)

                vert_var1_n1 = []
                vert_var2_n1 = []
                vert_var3_n1 = []
                vert_var4_n1 = []

                for cond_vert in range (len(coordinates)):
                    vert_var1_n1.append (vert_var1[cond_vert])
                    vert_var2_n1.append (vert_var2[cond_vert])
                    vert_var3_n1.append (vert_var3[cond_vert])
                    vert_var4_n1.append (vert_var4[cond_vert])
            
                vert_var1_n1 = np.asarray (vert_var1_n1)
                vert_var2_n1 = np.asarray (vert_var2_n1)
                vert_var3_n1 = np.asarray (vert_var3_n1)
                vert_var4_n1 = np.asarray (vert_var4_n1)

                # print ("var1_n1 = ", vert_var1_n1)

                # # Well-posedness check
                # i = 0
                # for var1_test, var2_test, var3_test, var4_test in zip (vert_var1_n1, vert_var2_n1, vert_var3_n1, vert_var4_n1):
                #     Aeval = np.asarray (Amat (var1_test, var2_test, var3_test, var4_test))
                #     Beval = np.asarray (Bmat (var1_test, var2_test, var3_test, var4_test))

                #     m_test, _ = eig (Beval, b = Aeval, overwrite_a = True, overwrite_b = True, check_finite = True)

                #     listreal = []
                #     listimag = []

                #     for cond0 in range (len (m_test)):
                #         realpart = m_test [cond0].real                
                #         imagpart = m_test [cond0].imag
                        
                #         listimag.append (imagpart)
                #         listreal.append (realpart)
                    
                #     if any ([ cond1 != 0 for cond1 in listimag ]) :
                #         print("INFO: ill-posed equation system")
                #         break
                #     else:
                #         i += 1
                #     break

                # # Condition for low liquid level of transition to intermittent flow
                # if any ([cond2 <= 0 for cond2 in vert_var1_n1]) :
                #     print("INFO: gas phase dominant. Low liquid level. Liquid equations vanish")
                #     break
                # elif any ([cond2 >= 1 for cond2 in vert_var1_n1]):
                #     print("INFO: liquid phase dominant. Intermittent flow. Gas equations vanish")
                #     break
                # else:
                #     pass
                # # Condition for positive velocities and pressure
                # if any ([ cond3 <= 0 for cond3 in vert_var2_n1]) :
                #     print("INFO: negative liquid velocity")
                #     break
                # else:
                #     pass

                # if any ([ cond4 <= 0 for cond4 in vert_var3_n1]) :
                #     print("INFO: negative gas velocity")
                #     break
                # else:
                #     pass

                # if any ([ cond5 <= 0 for cond5 in vert_var4_n1]) :
                #     print("INFO: negative pressure")
                #     break
                # else:
                #     pass
                
                # # Reynolds check
                # Rel_n1 = Re_l (vert_var1_n1, vert_var2_n1)
                # Reg_n1 = Re_g (vert_var1_n1, vert_var3_n1, vert_var4_n1)

                # if any ([ cond3 <= 1180 for cond3 in Rel_n1]):
                # # Rel < 1180:
                #     print("Liquid laminar")
                # elif  any ([ cond3 > 1180 for cond3 in Rel_n1]):
                #     pass

                # if any ([ cond4 <= 1180 for cond4 in Reg_n1]):
                #     print("Gas laminar")
                # elif  any ([ cond4 > 1180 for cond4 in Reg_n1]):
                #     pass
    # ===============================================================
    # CURRENT TIME BDF2
    # ===============================================================
            if time_method == 'BDF2':
                # Method
                # print('INFO: second-order Backward differentiation formula time integration.')

                # Initial conditions
                if t == 0:
                    (var1_past, var2_past, var3_past, var4_past) = var_past.split (deepcopy = True)
                    # Compute  nodal values (initial)
                    # n_val1_past = np.array (var1_past.vector ())
                    # n_val2_past = np.array (var2_past.vector ())
                    # n_val3_past = np.array (var3_past.vector ())
                    # n_val4_past = np.array (var4_past.vector ())

                    # nodal_var1_past = n_val1_past [::-1]
                    # nodal_var2_past = n_val2_past [::-1]
                    # nodal_var3_past = n_val3_past [::-1]
                    # nodal_var4_past = n_val4_past [::-1]

                    # # print("n", nodal_var1_n)

                    # nodes_var1_past = len (nodal_var1_n)
                    # nodes_var2_n = len (nodal_var2_n)
                    # nodes_var3_n = len (nodal_var3_n)
                    # nodes_var4_n = len (nodal_var4_n)

                    # print ("dof subspace var1 =", nodes_var1_n)
                    # print ("dof subspace var2 =", nodes_var2_n)
                    # print ("dof subspace var3 =", nodes_var3_n)
                    # print ("dof subspace var4 =", nodes_var4_n)

                    # Compute vertex values (initial step)
                    v_var1_past = var1_past.compute_vertex_values (mesh)
                    v_var2_past = var2_past.compute_vertex_values (mesh)
                    v_var3_past = var3_past.compute_vertex_values (mesh)
                    v_var4_past = var4_past.compute_vertex_values (mesh)

                    vert_var1_past = []
                    vert_var2_past = []
                    vert_var3_past = []
                    vert_var4_past = []

                    for cond_vert in range (len(coordinates)):
                        vert_var1_past.append (v_var1_past[cond_vert])
                        vert_var2_past.append (v_var2_past[cond_vert])
                        vert_var3_past.append (v_var3_past[cond_vert])
                        vert_var4_past.append (v_var4_past[cond_vert])
                    
                    vert_var1_past = np.asarray (vert_var1_past)
                    vert_var2_past = np.asarray (vert_var2_past)
                    vert_var3_past = np.asarray (vert_var3_past)
                    vert_var4_past = np.asarray (vert_var4_past)

                    # print("var1_past = ", vert_var1_past)

                    # # Well-posedness check
                    # i = 0
                    # for var1_test, var2_test, var3_test, var4_test in zip (vert_var1_n, vert_var2_n, vert_var3_n, vert_var4_n):
                    #     Aeval = np.asarray (Amat (var1_test, var2_test, var3_test, var4_test))
                    #     Beval = np.asarray (Bmat (var1_test, var2_test, var3_test, var4_test))

                    #     m_test, _ = eig (Beval, b = Aeval, overwrite_a = True, overwrite_b = True, check_finite = True)

                    #     listreal = []
                    #     listimag = []

                    #     for cond0 in range (len (m_test)):
                    #         realpart = m_test [cond0].real                
                    #         imagpart = m_test [cond0].imag
                            
                    #         listimag.append (imagpart)
                    #         listreal.append (realpart)
                        
                    #     if any ([ cond1 != 0 for cond1 in listimag ]) :
                    #         print("INFO: ill-posed equation system")
                    #         break
                    #     else:
                    #         i += 1
                    #     break

                    # # Condition for low liquid level of transition to intermittent flow
                    # if any ([cond2 <= 0 for cond2 in vert_var1_n]) :
                    #     print("INFO: gas phase dominant. Low liquid level. Liquid equations vanish")
                    #     break
                    # elif any ([cond2 >= 1 for cond2 in vert_var1_n]):
                    #     print("INFO: liquid phase dominant. Intermittent flow. Gas equations vanish")
                    #     break
                    # else:
                    #     pass
                    # # Condition for positive velocities and pressure
                    # if any ([ cond3 <= 0 for cond3 in vert_var2_n]) :
                    #     print("INFO: negative liquid velocity")
                    #     break
                    # else:
                    #     pass

                    # if any ([ cond4 <= 0 for cond4 in vert_var3_n]) :
                    #     print("INFO: negative gas velocity")
                    #     break
                    # else:
                    #     pass

                    # if any ([ cond5 <= 0 for cond5 in vert_var4_n]) :
                    #     print("INFO: negative pressure")
                    #     break
                    # else:
                    #     pass
                    
                    # # Reynolds check
                    # Rel_n = Re_l (vert_var1_n, vert_var2_n)
                    # Reg_n = Re_g (vert_var1_n, vert_var3_n, vert_var4_n)

                    # if any ([ cond3 <= 1180 for cond3 in Rel_n]):
                    # # Rel < 1180:
                    #     print("Liquid laminar")
                    # elif  any ([ cond3 > 1180 for cond3 in Rel_n]):
                    #     pass

                    # if any ([ cond4 <= 1180 for cond4 in Reg_n]):
                    #     print("Gas laminar")
                    # elif  any ([ cond4 > 1180 for cond4 in Reg_n]):
                    #     pass

                    # c = conditional (ge (var1, 1), 2, info (var1, verbose=False))
                    # c = conditional (le (var1, 0), str(INFO: liquid phase dominant, which leads to intermittent flow. Gas equations vanish), 1)

                    # # Plot solution
                    # plt.figure (5, figsize = (8, 4))
                    # plt.xlim (0, l)
                    # plt.ylim (0, 1)
                    # plt.grid (True, which = "both")
                    # plot (var1_past, title = r"$\alpha_l$", rescale = True)

                    # plt.figure (6, figsize = (8, 4))
                    # plt.xlim (0, l)
                    # plt.ylim (min(vert_var2_past), max(vert_var2_past))
                    # plt.grid (True, which = "both")
                    # plot (var2_past, title = r"$u_l$", rescale = True)

                    # plt.figure (7, figsize = (8, 4))
                    # plt.xlim (0, l)
                    # plt.grid (True, which = "both")
                    # plot (var3_past, title = r"$u_g$", rescale = True)

                    # plt.figure (8, figsize = (8, 4))
                    # plt.xlim (0, l)
                    # plt.grid (True, which = "both")
                    # plot (var4_past, title = r"$p_i$", rescale = True)
                    
                    # Save solution
                    # ff_var1 << var1_past
                    # ff_var2 << var2_past
                    # ff_var3 << var3_past
                    # ff_var4 << var4_past

                    # Time stepping
                    t = dt/2
                                    
                    # Parameters
                    a0 = 1
                    a1 = -1
                    a2 = 0
                    theta = 1
                    
                    # Amat
                    Am1 = ( (a0*var1_n + a1*var1_past)/dt*v1 )*dx 
                    Am2 = ( (a0*var2_n + a1*var1_past)/dt*v2 )*dx 
                    Am3 = ( (a0*var3_n + a1*var1_past)/dt*v3 )*dx 
                    Am4 = ( (a0*var4_n + a1*var1_past)/dt*v4 )*dx

                    # Variational form
                    F_n = visc*(Cm1_lin_n + Cm2_lin_n + Cm3_lin_n + Cm4_lin_n) - (Bm1_n + Bm2_n + Bm3_n + Bm4_n)
                    F_past = visc*(Cm1_lin_past + Cm2_lin_past + Cm3_lin_past + Cm4_lin_past) - (Bm1_past + Bm2_past + Bm3_past + Bm4_past)
                    R = (Am1 + Am2 + Am3 + Am4) - theta*F_n - (1 - theta)*F_past 

                    # Compute directional derivative about u in the direction of du (Jacobian)
                    dF = derivative (R, var_n, dvar)

                    problem = NonlinearVariationalProblem (R, var_n, bcs = bcs, J = dF, form_compiler_parameters = ffc_options)
                    solver  = NonlinearVariationalSolver (problem)
                    prm     = solver.parameters
                    info (prm, True)
                    
                    # Nonlinear solver
                    prm ['nonlinear_solver'] = 'newton'
                    # prm ['print_matrix']     = False #True
                    # prm ['print_rhs']        = False #True
                    # prm ['symmetric']        = False #True
                    
                    # Linear solver
                    # prm ['newton_solver']['absolute_tolerance']      = 1e-1 #1E-8
                    # prm ['newton_solver']['convergence_criterion']   = 'residual' #'residual' 'incremental'
                    # prm ['newton_solver']['error_on_nonconvergence'] = True
                    prm ['newton_solver']['linear_solver']           = 'umfpack' # 'bicgstab' 'cg' 'gmres' 'minres' 'petsc' 'richardson' 'superlu_dist' 'tfqmr' 'umfpack'
                    # prm ['newton_solver']['maximum_iterations']      = 10000
                    # prm ['newton_solver']['preconditioner']          = 'ilu' # 'ilu' 'icc' 'petsc_amg' 'sor'
                    # prm ['newton_solver']['relative_tolerance']      = 1e-1
                    # prm ['newton_solver']['relaxation_parameter']    = 1.0
                    # prm ['newton_solver']['report']                  = True

                    # Krylov solver
                    # prm ['newton_solver']['krylov_solver']['absolute_tolerance']       = 1e-3 #1E-9
                    # #     prm ['newton_solver']['krylov_solver']['error_on_nonconvergence']  = True
                    # prm ['newton_solver']['krylov_solver']['maximum_iterations']       = 10000 # 500000
                    # prm ['newton_solver']['krylov_solver']["monitor_convergence"]      = True
                    # prm ['newton_solver']['krylov_solver']["nonzero_initial_guess"]    = True #False
                    # prm ['newton_solver']['krylov_solver']['relative_tolerance']       = 1e-3
                    # prm ['newton_solver']['krylov_solver']['report']                   = True
                        
                    # LU solver
                    # prm ['newton_solver']['lu_solver']['report']    = True
                    # prm ['newton_solver']['lu_solver']['symmetric'] = False
                    # prm ['newton_solver']['lu_solver']['verbose']   = True

                    solver.solve ()  

                    # Split of the vector var
                    (var1_n, var2_n, var3_n, var4_n) = var_n.split (deepcopy = True)

                    # L2 norm
                    # infonorm = var1.vector().norm("l2")
                    # print("Solution vector norm (0): {!r}".format(infonorm)) 
                    # L2norm_var1.append (var1.vector().norm("l2"))
                    # L2norm_var2.append (var2.vector().norm("l2"))   
                    # L2norm_var3.append (var3.vector().norm("l2"))   
                    # L2norm_var4.append (var4.vector().norm("l2"))   

                    # Nodal values (current)
                    # n_val1_n1 = np.array(var1.vector())
                    # n_val2_n1 = np.array(var2.vector())
                    # n_val3_n1 = np.array(var3.vector())
                    # n_val4_n1 = np.array(var4.vector())

                    # nodal_var1_n1 = n_val1_n1 [::-1]
                    # nodal_var2_n1 = n_val2_n1 [::-1]
                    # nodal_var3_n1 = n_val3_n1 [::-1]
                    # nodal_var4_n1 = n_val4_n1 [::-1]

                    # # print("n+1", nodal_var1_n1)

                    # # Compute vertex values (next step)
                    # vert_var1 = var1.compute_vertex_values (mesh)
                    # vert_var2 = var2.compute_vertex_values (mesh)
                    # vert_var3 = var3.compute_vertex_values (mesh)
                    # vert_var4 = var4.compute_vertex_values (mesh)

                    # vert_var1_n1 = []
                    # vert_var2_n1 = []
                    # vert_var3_n1 = []
                    # vert_var4_n1 = []

                    # for cond_vert in range (len(coordinates)):
                    #     vert_var1_n1.append (vert_var1[cond_vert])
                    #     vert_var2_n1.append (vert_var2[cond_vert])
                    #     vert_var3_n1.append (vert_var3[cond_vert])
                    #     vert_var4_n1.append (vert_var4[cond_vert])
                
                    # vert_var1_n1 = np.asarray (vert_var1_n1)
                    # vert_var2_n1 = np.asarray (vert_var2_n1)
                    # vert_var3_n1 = np.asarray (vert_var3_n1)
                    # vert_var4_n1 = np.asarray (vert_var4_n1)

                    # print ("var1_n1 = ", vert_var1_n1)

                    # # Well-posedness check
                    # i = 0
                    # for var1_test, var2_test, var3_test, var4_test in zip (vert_var1_n1, vert_var2_n1, vert_var3_n1, vert_var4_n1):
                    #     Aeval = np.asarray (Amat (var1_test, var2_test, var3_test, var4_test))
                    #     Beval = np.asarray (Bmat (var1_test, var2_test, var3_test, var4_test))

                    #     m_test, _ = eig (Beval, b = Aeval, overwrite_a = True, overwrite_b = True, check_finite = True)

                    #     listreal = []
                    #     listimag = []

                    #     for cond0 in range (len (m_test)):
                    #         realpart = m_test [cond0].real                
                    #         imagpart = m_test [cond0].imag
                            
                    #         listimag.append (imagpart)
                    #         listreal.append (realpart)
                        
                    #     if any ([ cond1 != 0 for cond1 in listimag ]) :
                    #         print("INFO: ill-posed equation system")
                    #         break
                    #     else:
                    #         i += 1
                    #     break

                    # # Condition for low liquid level of transition to intermittent flow
                    # if any ([cond2 <= 0 for cond2 in vert_var1_n1]) :
                    #     print("INFO: gas phase dominant. Low liquid level. Liquid equations vanish")
                    #     break
                    # elif any ([cond2 >= 1 for cond2 in vert_var1_n1]):
                    #     print("INFO: liquid phase dominant. Intermittent flow. Gas equations vanish")
                    #     break
                    # else:
                    #     pass
                    # # Condition for positive velocities and pressure
                    # if any ([ cond3 <= 0 for cond3 in vert_var2_n1]) :
                    #     print("INFO: negative liquid velocity")
                    #     break
                    # else:
                    #     pass

                    # if any ([ cond4 <= 0 for cond4 in vert_var3_n1]) :
                    #     print("INFO: negative gas velocity")
                    #     break
                    # else:
                    #     pass

                    # if any ([ cond5 <= 0 for cond5 in vert_var4_n1]) :
                    #     print("INFO: negative pressure")
                    #     break
                    # else:
                    #     pass
                    
                    # # Reynolds check
                    # Rel_n1 = Re_l (vert_var1_n1, vert_var2_n1)
                    # Reg_n1 = Re_g (vert_var1_n1, vert_var3_n1, vert_var4_n1)

                    # if any ([ cond3 <= 1180 for cond3 in Rel_n1]):
                    # # Rel < 1180:
                    #     print("Liquid laminar")
                    # elif  any ([ cond3 > 1180 for cond3 in Rel_n1]):
                    #     pass

                    # if any ([ cond4 <= 1180 for cond4 in Reg_n1]):
                    #     print("Gas laminar")
                    # elif  any ([ cond4 > 1180 for cond4 in Reg_n1]):
                    #     pass
                    
                    t = 0
                else:
                    pass

                # Time stepping
                t = dt
                
                # Print progress
                # clear_output ()
                # print ("Iteration :", int (round (t/dt)), "of", num_steps)
                # print ("Time      :", t, "s")
                                
                # Parameters
                a0 = 3/2
                a1 = -2
                a2 = 1/2
                theta = 1
                
                # Amat
                Am1 = ( (a0*var1 + a1*var1_n + a2*var1_past)/dt*v1 )*dx 
                Am2 = ( (a0*var2 + a1*var2_n + a2*var1_past)/dt*v2 )*dx 
                Am3 = ( (a0*var3 + a1*var3_n + a2*var1_past)/dt*v3 )*dx 
                Am4 = ( (a0*var4 + a1*var4_n + a2*var1_past)/dt*v4 )*dx

                # Variational form
                F = visc*(Cm1_lin + Cm2_lin + Cm3_lin + Cm4_lin) - (Bm1 + Bm2 + Bm3 + Bm4)
                F_n = visc*(Cm1_lin_n + Cm2_lin_n + Cm3_lin_n + Cm4_lin_n) - (Bm1_n + Bm2_n + Bm3_n + Bm4_n)
                R = (Am1 + Am2 + Am3 + Am4) - theta*F - (1 - theta)*F_n 

                # Compute directional derivative about u in the direction of du (Jacobian)
                dF = derivative (R, var, dvar)

                problem = NonlinearVariationalProblem (R, var, bcs = bcs, J = dF, form_compiler_parameters = ffc_options)
                solver  = NonlinearVariationalSolver (problem)
                prm     = solver.parameters
                info (prm, True)
                
                # Nonlinear solver
                prm ['nonlinear_solver'] = 'newton'
                prm ['print_matrix']     = False #True
                prm ['print_rhs']        = False #True
                prm ['symmetric']        = False #True
                
                # Linear solver
                # prm ['newton_solver']['absolute_tolerance']      = 1e-1 #1E-8
                # prm ['newton_solver']['convergence_criterion']   = 'residual' #'residual' 'incremental'
                # prm ['newton_solver']['error_on_nonconvergence'] = True
                prm ['newton_solver']['linear_solver']           = 'umfpack' # 'bicgstab' 'cg' 'gmres' 'minres' 'petsc' 'richardson' 'superlu_dist' 'tfqmr' 'umfpack'
                # prm ['newton_solver']['maximum_iterations']      = 10000
                # prm ['newton_solver']['preconditioner']          = 'ilu' # 'ilu' 'icc' 'petsc_amg' 'sor'
                # prm ['newton_solver']['relative_tolerance']      = 1e-1
                # prm ['newton_solver']['relaxation_parameter']    = 1.0
                # prm ['newton_solver']['report']                  = True

                # Krylov solver
                # prm ['newton_solver']['krylov_solver']['absolute_tolerance']       = 1e-3 #1E-9
                # #     prm ['newton_solver']['krylov_solver']['error_on_nonconvergence']  = True
                # prm ['newton_solver']['krylov_solver']['maximum_iterations']       = 10000 # 500000
                # prm ['newton_solver']['krylov_solver']["monitor_convergence"]      = True
                # prm ['newton_solver']['krylov_solver']["nonzero_initial_guess"]    = True #False
                # prm ['newton_solver']['krylov_solver']['relative_tolerance']       = 1e-3
                # prm ['newton_solver']['krylov_solver']['report']                   = True
                    
                # LU solver
                # prm ['newton_solver']['lu_solver']['report']    = True
                # prm ['newton_solver']['lu_solver']['symmetric'] = False
                # prm ['newton_solver']['lu_solver']['verbose']   = True

                solver.solve ()  

                # Split of the vector var
                (var1, var2, var3, var4) = var.split (deepcopy = True)

                # L2 norm
                # infonorm = var1.vector().norm("l2")
                # print("Solution vector norm (0): {!r}".format(infonorm)) 
                L2norm_var1.append (var1.vector().norm("l2"))
                L2norm_var2.append (var2.vector().norm("l2"))   
                L2norm_var3.append (var3.vector().norm("l2"))   
                L2norm_var4.append (var4.vector().norm("l2"))   

                # Nodal values (current)
                n_val1_n1 = np.array(var1.vector())
                n_val2_n1 = np.array(var2.vector())
                n_val3_n1 = np.array(var3.vector())
                n_val4_n1 = np.array(var4.vector())

                nodal_var1_n1 = n_val1_n1 [::-1]
                nodal_var2_n1 = n_val2_n1 [::-1]
                nodal_var3_n1 = n_val3_n1 [::-1]
                nodal_var4_n1 = n_val4_n1 [::-1]

                # print("n+1", nodal_var1_n1)

                # Compute vertex values (next step)
                vert_var1 = var1.compute_vertex_values (mesh)
                vert_var2 = var2.compute_vertex_values (mesh)
                vert_var3 = var3.compute_vertex_values (mesh)
                vert_var4 = var4.compute_vertex_values (mesh)

                vert_var1_n1 = []
                vert_var2_n1 = []
                vert_var3_n1 = []
                vert_var4_n1 = []

                for cond_vert in range (len(coordinates)):
                    vert_var1_n1.append (vert_var1[cond_vert])
                    vert_var2_n1.append (vert_var2[cond_vert])
                    vert_var3_n1.append (vert_var3[cond_vert])
                    vert_var4_n1.append (vert_var4[cond_vert])
            
                vert_var1_n1 = np.asarray (vert_var1_n1)
                vert_var2_n1 = np.asarray (vert_var2_n1)
                vert_var3_n1 = np.asarray (vert_var3_n1)
                vert_var4_n1 = np.asarray (vert_var4_n1)

                # print ("var1_n1 = ", vert_var1_n1)

                # # Well-posedness check
                # i = 0
                # for var1_test, var2_test, var3_test, var4_test in zip (vert_var1_n1, vert_var2_n1, vert_var3_n1, vert_var4_n1):
                #     Aeval = np.asarray (Amat (var1_test, var2_test, var3_test, var4_test))
                #     Beval = np.asarray (Bmat (var1_test, var2_test, var3_test, var4_test))

                #     m_test, _ = eig (Beval, b = Aeval, overwrite_a = True, overwrite_b = True, check_finite = True)

                #     listreal = []
                #     listimag = []

                #     for cond0 in range (len (m_test)):
                #         realpart = m_test [cond0].real                
                #         imagpart = m_test [cond0].imag
                        
                #         listimag.append (imagpart)
                #         listreal.append (realpart)
                    
                #     if any ([ cond1 != 0 for cond1 in listimag ]) :
                #         print("INFO: ill-posed equation system")
                #         break
                #     else:
                #         i += 1
                #     break

                # # Condition for low liquid level of transition to intermittent flow
                # if any ([cond2 <= 0 for cond2 in vert_var1_n1]) :
                #     print("INFO: gas phase dominant. Low liquid level. Liquid equations vanish")
                #     break
                # elif any ([cond2 >= 1 for cond2 in vert_var1_n1]):
                #     print("INFO: liquid phase dominant. Intermittent flow. Gas equations vanish")
                #     break
                # else:
                #     pass
                # # Condition for positive velocities and pressure
                # if any ([ cond3 <= 0 for cond3 in vert_var2_n1]) :
                #     print("INFO: negative liquid velocity")
                #     break
                # else:
                #     pass

                # if any ([ cond4 <= 0 for cond4 in vert_var3_n1]) :
                #     print("INFO: negative gas velocity")
                #     break
                # else:
                #     pass

                # if any ([ cond5 <= 0 for cond5 in vert_var4_n1]) :
                #     print("INFO: negative pressure")
                #     break
                # else:
                #     pass
                
                # # Reynolds check
                # Rel_n1 = Re_l (vert_var1_n1, vert_var2_n1)
                # Reg_n1 = Re_g (vert_var1_n1, vert_var3_n1, vert_var4_n1)

                # if any ([ cond3 <= 1180 for cond3 in Rel_n1]):
                # # Rel < 1180:
                #     print("Liquid laminar")
                # elif  any ([ cond3 > 1180 for cond3 in Rel_n1]):
                #     pass

                # if any ([ cond4 <= 1180 for cond4 in Reg_n1]):
                #     print("Gas laminar")
                # elif  any ([ cond4 > 1180 for cond4 in Reg_n1]):
                #     pass

# ===============================================================
# END OF COMPUTATION FOR ALL TIME METHODS
# ===============================================================
            if time_method == 'BDF2':
                var_past.vector ()[:] = var_n.vector () 

            var_n.vector ()[:] = var.vector () 

# ===============================================================
# EIGENSPECTRUM
# ===============================================================
            maprealeig = [[ ] for iiiii in range (len (vert_var1_n1))]
            mapimageig = [[ ] for iiiii in range (len (vert_var1_n1))]

            list_unstable = []
            
            iiiiii = 0
            for var1_test, var2_test, var3_test, var4_test in zip (vert_var1_n1, vert_var2_n1, vert_var3_n1, vert_var4_n1):
                Aeval = np.asarray (Amat (var1_test, var2_test, var3_test, var4_test))
                Beval = np.asarray (Bmat (var1_test, var2_test, var3_test, var4_test))
                Ceval = np.asarray (Cmat_lin (var1_test, var2_test, var3_test, var4_test))

                Acomplex   = Aeval.dot (1j)
                Bcomplex   = Beval.dot (1j)
                
                m_eig, _ = eig ( (-wavenumber*Bcomplex - Ceval), b =  -Acomplex, overwrite_a = True, overwrite_b = True, check_finite = True) 

                listreal_eig = []
                listimag_eig = []

                for cond8 in range (len (m_eig)):
                    realpart_eig = m_eig [cond8].real
                    imagpart_eig = m_eig [cond8].imag
                    
                    listimag_eig.append (imagpart_eig)
                    listreal_eig.append (realpart_eig)
                
                    mapimageig[iiiiii] = np.array(listimag_eig)
                    maprealeig[iiiiii] = np.array(listreal_eig)

                if all ([ cond11 > 0 for cond11 in mapimageig[iiiiii]]):
                    # print("nao imprimir")
                    co = 'stable1'
                    iiiiii += 1
                elif any ([ cond11 < 0 for cond11 in mapimageig[iiiiii]]):
                    print("Unstable")
                    print ("este valoooor = ", j_l)
                    co = 'unstable1'
                    j_lmapstability [k, j] = j_l
                    break

            if co == 'unstable1':
                break
            else:
                i += 1
            
            del var1
            del var2
            del var3
            del var4

        j += 1
    k += 1
```
